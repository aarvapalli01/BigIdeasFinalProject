# Addressing the Avian Housing Crisis



```python
import pandas as pd
```

```python
trees = pd.read_csv("trees.csv")
trees.head()
```

```python
items = dict()
for index, row in trees.iterrows():
    if row['neighborhood'] not in items:
        items[row['neighborhood']] = 1
    else:
        items[row['neighborhood']] += 1
```

```python
items
```

```python
sortedNbhds = sorted(items.items(),key=lambda x:x[1],reverse=True)
sortedNbhds
```
